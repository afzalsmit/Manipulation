```python
price_map = {
    'Software': 350,
    'Hardware': 425,
    'Service': 275
}
df['Revenue'] = df['Product'].map(price_map) * df['Units']
```

```python
df.index = pd.to_datetime(df.index)

# Extract day of the week from the index and store it in a new column
df['DayOfWeek'] = df.index.day_name()

print(df[['DayOfWeek']].head())
```

```python
# 14. Apply a discount function to the Revenue column
def apply_discount(row):
    if row['Units'] >= 20:
        return row['Revenue'] * 0.90  # 10% discount
    elif row['Units'] >= 15:
        return row['Revenue'] * 0.95  # 5% discount
    else:
        return row['Revenue']

df['DiscountedRevenue'] = df.apply(apply_discount, axis=1)
```
